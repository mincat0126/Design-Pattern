````python
from os import X_OK
from ast import match_case
import time
from random import *
from google.colab import output
from abc import ABC, abstractmethod



# =========================
# Character
# =========================

class Character():
  def __init__(self,health):
    self.health=health

  def isAlive(self):
    return self.health>0

  def takeDamage(self,damage):
    self.health-=damage
    if self.health<0:
      self.health=0


class Player(Character):
  def __init__(self):
    super().__init__(health=15)

class Enemy(Character):
  def __init__(self):
    super().__init__(health=randint(10,20))



# =========================
# AttackAction
# =========================


class AttackActionAbstract(ABC):
  @abstractmethod
  def playAction(self,enemy):
    pass


class AttackAction(AttackActionAbstract):
  def playAction(self,enemy):
    damage=randint(1,3)
    enemy.takeDamage(damage)
    print(f"Player attack {damage} damage!")


# =========================
# DefenseAction
# =========================



class DefenseActionAbstract(ABC):
  @abstractmethod
  def defenseAction(self,player,enemy,damage):
    pass


class CounterDefense(DefenseActionAbstract):
  def defenseAction(self,player,enemy,enemyDamage):
    blockAmount=randint(0,2)
    print(f"Player take {enemyDamage} damage and block {blockAmount} damage!")
    enemyDamage=max(0,enemyDamage-blockAmount)
    player.takeDamage(enemyDamage)
    time.sleep(1)

    if(player.isAlive()):
      playerDamage=randint(1,2)
      enemy.takeDamage(playerDamage)
      print(f"Player counter {playerDamage} damage!")

class AvoidDefense(DefenseActionAbstract):
  def defenseAction(self,player,enemy,damage):
    isAvoid=randint(0,1)
    if(isAvoid==1):
      damage=2
      print(f"Avoid success! and attack {damage} damage!")
      enemy.takeDamage(damage)
    else:
      print(f"Avoid fail! and take {damage} damage!")
      player.takeDamage(damage)


# =========================
# InputManager
# =========================


class InputManager():

  def __init__(self):
    self.key=""

  def getKey(self):
    self.key=input().upper()
    return self.key

  def checkKey(self,key):
    match self.key:
      case 'Z':
        return "attack"
      case 'X':
        return "counter"
      case 'C':
        return "avoid"
      case _:
        print("Error: please enter the correct key")
        return "-1"

# =========================
# PrintText
# =========================

class PrintText():
  def printInfo(self,enemy,player):
    print("\n\n\n")
    print(f"Enemy Health: {enemy.health}, Player Health: {player.health}")
    print("attack: Z(1 ~ 3)   counter: X(0 ~ 2 block, 1~2 attack)   avoid: C(50% avoid and 2 attack)")
    print("-"*30)

  def printResult(self,isWin):
    if(isWin==True):
      print("Win!")
    else:
      print("Game Over!")

# =========================
# checkAction()
# =========================

class CheckAction():
  def __init__(self,player,enemy):
    self.player=player
    self.enemy=enemy

    self.attackType={
    "attack":AttackAction(),
    }

    self.defenseType={
    "counter":CounterDefense(),
    "avoid":AvoidDefense()
    }

  def isAttack(self,actionType):
    if(actionType in self.attackType):
      attackAction=self.attackType[actionType]

      attackAction.playAction(self.enemy)
      time.sleep(1)

  def isDefense(self,actionType):
    damage=randint(1,3)
    if(actionType in self.defenseType):
      defenseAction=self.defenseType[actionType]
      defenseAction.defenseAction(self.player,self.enemy,damage)
    else:
      self.player.takeDamage(damage)
      print(f"Player take {damage} damage!")


# =========================
# GameManager
# =========================


class GameManager():
  def __init__(self):
    self.player=Player()
    self.enemy=Enemy()
    self.printText=PrintText()
    self.inputManager=InputManager()
    self.checkAction=CheckAction(self.player,self.enemy)

  def playerTurn(self):
    key="-1"
    while key== "-1":
      self.printText.printInfo(self.enemy,self.player)
      key=self.inputManager.checkKey(self.inputManager.getKey())

    self.checkAction.isAttack(key)
    if(self.enemy.isAlive()):
      self.checkAction.isDefense(key)


  def run(self):
    while self.player.isAlive() and self.enemy.isAlive():
      self.playerTurn()

    if self.player.isAlive():
      isWin=True
      self.printText.printResult(isWin)
    else:
      isWin=False
      self.printText.printResult(isWin)

# =========================
# start
# ====================z=====

def start():
  game=GameManager()
  game.run()

start()
````

````python
from abc import ABC, abstractmethod
from random import *

# =========================
# CapeType
# =========================

class CapeType(ABC):
  @abstractmethod
  def getCape(self):
    pass

class KnightCape(CapeType):
  def getCape(self):
    print("get Knight cape")

class ArcherCape(CapeType):
  def getCape(self):
    print("get Archer cape")

# =========================
# WeaponType
# =========================

class SwordType(ABC):
  def weaponInfo(self):
    pass

class Dagger(SwordType):
  def __init__(self):
    self.damage=1

  def weaponInfo(self):
    print(f"damage: {self.damage}")

class LongSword(SwordType):
  def __init__(self):
    self.damage=5

  def weaponInfo(self):
    print(f"damage: {self.damage}")


class BowType(ABC):
  def weaponInfo(self):
    pass

class ShortBow(BowType):
  def __init__(self):
    self.range=2

  def weaponInfo(self):
    print(f"range: {self.range}")

class LongBow(BowType):
  def __init__(self):
    self.range=5

  def weaponInfo(self):
    print(f"range: {self.range}")

# =========================
# WeaponFactory
# =========================

class WeaponFactory(ABC):
  @abstractmethod
  def createWeapon(self):
    pass

  @abstractmethod
  def createCape(self):
    pass


class SwordFactory(WeaponFactory):
  def createWeapon(self):
    self.swordType=randint(0,1)
    if(self.swordType==0):
      return Dagger()
    else:
      return LongSword()

  def createCape(self):
    return KnightCape()

class BowFactory(WeaponFactory):
  def createWeapon(self):
    self.bowType=randint(0,1)
    if(self.bowType==0):
      return ShortBow()
    else:
      return LongBow()

  def createCape(self):
    return ArcherCape()

# =========================
# test
# =========================

def main():
  swordFactory=SwordFactory()
  sword=swordFactory.createWeapon()
  swordCape=swordFactory.createCape()
  sword.weaponInfo()
  swordCape.getCape()
  print("\n")

  bowFactory=BowFactory()
  bow=bowFactory.createWeapon()
  bowCape=bowFactory.createCape()
  bow.weaponInfo()
  bowCape.getCape()

main()
````

````python
from abc import ABC, abstractmethod

# =========================
# Humidity
# =========================

class Humidity(ABC):
  @abstractmethod
  def humidityInfo(self):
    pass

class Normal(Humidity):
  def __init__(self):
    self.humidity=randint(45,65)

  def humidityInfo(self):
    print(f"Normal humidity! humidity: {self.humidity}%")

#ì´‰ì´‰í•œ
class Moist(Humidity):
  def __init__(self):
    self.humidity=randint(70,82)

  def humidityInfo(self):
    print(f"Just Moist! humidity: {self.humidity}%")

# ìŠµê¸°ì°¬
class Dank(Humidity):
  def __init__(self):
    self.humidity=randint(85,99)

  def humidityInfo(self):
    print(f"Very damp! humidity: {self.humidity}%")

# =========================
# Cloud
# =========================

class Cloud(ABC):
  @abstractmethod
  def setCloud():
    pass

class ClearCloud(Cloud):
  def setCloud(self):
    print("Clear Cloud!")

class RainCloud(Cloud):
  def setCloud(self):
    print("Rain Cloud!")

class DarkCloud(Cloud):
  def setCloud(self):
    print("Dark Cloud!")

# =========================
# Weather
# =========================

class WeatherFactory(ABC):
  @abstractmethod
  def cloudState(self):
    pass

  @abstractmethod
  def humidityState(self):
    pass

class SunnyFactory(WeatherFactory):
  def cloudState(self):
    return ClearCloud()

  def humidityState(self):
    return Normal()

class RainFactory(WeatherFactory):
  def cloudState(self):
    return RainCloud()

  def humidityState(self):
    return Moist()

class StromFactory(WeatherFactory):
  def cloudState(self):
    return DarkCloud()

  def humidityState(self):
    return Dank()

# =========================
# main
# =========================

def main():
  sunnyFactory=SunnyFactory()
  sunnyCloud=sunnyFactory.cloudState()
  sunnyCloud.setCloud()
  sunnyHumidity=sunnyFactory.humidityState()
  sunnyHumidity.humidityInfo()
  print("\n")

  rainFactory=RainFactory()
  rainCloud=rainFactory.cloudState()
  rainCloud.setCloud()
  rainHumidity=rainFactory.humidityState()
  rainHumidity.humidityInfo()
  print("\n")

  stormFactory=StromFactory()
  stormCloud=stormFactory.cloudState()
  stormCloud.setCloud()
  stormHumidity=stormFactory.humidityState()
  stormHumidity.humidityInfo()

main()
````

````python
from abc import ABC, abstractmethod

# =========================
# Restaurant
# =========================

class Restaurant(ABC):
  @abstractmethod
  def setTable(self):
    pass

class MeatRestaurant(Restaurant):
  def setTable(self):
    print("Setting Table\n")

class Buffet(Restaurant): #ë·”íŽ˜
  def setTable(self):
    print("Setting Table\n")

# =========================
# RestaurantProxy
# =========================

class RestaurantProxy(Restaurant):
  def __init__(self,Restaurant:Restaurant):
    self.restaurant=Restaurant
    self.remainTable=2

  def checkTable(self):
    if self.remainTable<=0:
      print("No more table\n")
      return
    else:
      self.remainTable-=1
      print(f"remain table: {self.remainTable}")
      self.setTable()

  def setTable(self):
    self.restaurant.setTable()

# =========================
# main
# =========================

def main():
  meatRestaurant=MeatRestaurant()
  meatRestaurantProxy=RestaurantProxy(meatRestaurant)
  meatRestaurantProxy.checkTable()
  meatRestaurantProxy.checkTable()
  meatRestaurantProxy.checkTable()

  buffet=Buffet()
  buffetProxy=RestaurantProxy(buffet)
  buffetProxy.checkTable()
  buffetProxy.checkTable()
  buffetProxy.checkTable()

main()
````

````python
import copy
import random
from abc import ABC, abstractmethod
from ast import match_case
import time

# =========================
# Block
# =========================

class Block():
    def __init__(self, value):
        self.value = int(value)

    def getValue(self):
        return self.value

# =========================
# BlockDecorator
# =========================

class BlockDecorator(Block):
    def __init__(self, block: Block):
        self.block = block

    def getValue(self):
        return self.block.getValue()

# =========================
# BonusBlock
# =========================

class BonusBlock(BlockDecorator):
    def __init__(self, block: Block):
        super().__init__(block)
        self.multiplier = 2

    def getValue(self):
        return self.block.getValue() * self.multiplier

# =========================
# MergeBlock
# =========================

class MergeBlock(Block):
    def __init__(self, value):
        self.value = value
        self.block = None

    def getBlock(self):
        if self.block is None:
            self.block = Block(self.value)
        return self.block

# =========================
# InputManager
# =========================

class InputManager():
    def __init__(self, puzzleManagerProxy):
        self.puzzleManagerProxy = puzzleManagerProxy

    def getKey(self):
        key = input("Move(w/a/s/d), Quit (q): ").lower()
        match key:
            case "w":
                self.puzzleManagerProxy.blockMove("up")
            case "a":
                self.puzzleManagerProxy.blockMove("left")
            case "s":
                self.puzzleManagerProxy.blockMove("down")
            case "d":
                self.puzzleManagerProxy.blockMove("right")
            case "q":
                print("Exiting...")
                time.sleep(2)
                raise SystemExit
            case _:
                print("Invalid input")

# =========================
# PuzzleManager
# =========================

class PuzzleManager():
    def __init__(self, size, grid):
        self.size = size
        self.grid = grid

    def spawnBlock(self):
        self.emptyPosition = []

        for j in range(0, self.size):
            for i in range(0, self.size):
                if self.grid[i][j] is None:
                    self.emptyPosition.append((i, j))

        i, j = random.choice(self.emptyPosition)
        self.grid[i][j] = Block(random.choice([2, 4]))

    def blockMove(self, dir):
        blockMoved = False
        for j in range(0, self.size):
            if dir in ("left", "right"):
                temp = []
                for i in range(self.size):
                    if self.grid[j][i] is not None:
                        temp.append(self.grid[j][i])
            elif dir in ("up", "down"):
                temp = []
                for i in range(self.size):
                    if self.grid[i][j] is not None:
                        temp.append(self.grid[i][j])

            if dir in ("right", "down"):
                temp.reverse()

            mergedValue = []
            i = 0
            while i < len(temp):
                if i + 1 < len(temp) and temp[i].getValue() == temp[i + 1].getValue():
                    mergedBlock = MergeBlock(temp[i].getValue() * 2).getBlock()
                    if random.randint(0, 9) == 1:
                        mergedBlock = BonusBlock(mergedBlock)
                        print(f"Bonus! Block doubled to {mergedBlock.getValue()}")
                    mergedValue.append(mergedBlock)
                    i += 2
                    blockMoved = True
                else:
                    mergedValue.append(temp[i])
                    i += 1

            while len(mergedValue) < self.size:
                mergedValue.append(None)

            if dir in ("right", "down"):
                mergedValue.reverse()

            for i in range(self.size):
                if dir in ("left", "right"):
                    if self.grid[j][i] != mergedValue[i]:
                        blockMoved = True
                    self.grid[j][i] = mergedValue[i]
                elif dir in ("up", "down"):
                    if self.grid[i][j] != mergedValue[i]:
                        blockMoved = True
                    self.grid[i][j] = mergedValue[i]

        if blockMoved:
            self.spawnBlock()

    def isGameOver(self):
        for i in range(self.size):
            for j in range(self.size):
                if self.grid[i][j] is None:
                    return False
                if i + 1 < self.size and self.grid[i + 1][j] != None and self.grid[i][j].getValue() == self.grid[i + 1][j].getValue():
                    return False
                if j + 1 < self.size and self.grid[i][j + 1] != None and self.grid[i][j].getValue() == self.grid[i][j + 1].getValue():
                    return False
        return True

# =========================
# PuzzleManagerProxy
# =========================

class PuzzleManagerProxy():
    def __init__(self, puzzleManager: PuzzleManager):
        self.puzzleManager = puzzleManager

    def spawnBlock(self):
        return self.puzzleManager.spawnBlock()

    def blockMove(self, dir):
        return self.puzzleManager.blockMove(dir)

    def isGameOver(self):
        return self.puzzleManager.isGameOver()

# =========================
# PrintText
# =========================

class PrintText():
    def __init__(self, size, grid):
        self.size = size
        self.grid = grid

    def printGrid(self):
        print("\n")
        for i in range(0, self.size):
            for j in range(0, self.size):
                if self.grid[i][j] is None:
                    print(" â–  ", end="")
                else:
                    print(f" {self.grid[i][j].getValue()} ", end="")
            print("\n")

    def gameOver(self):
        print("Game Over!")

# =========================
# GameManager
# =========================

class GameManager():
    def __init__(self):
        self.size = 2
        self.grid = []
        self.isGameOver = False
        for i in range(0, self.size):
            self.grid.append([None] * self.size)

        self.printText = PrintText(self.size, self.grid)
        self.puzzleManager = PuzzleManager(self.size, self.grid)
        self.puzzleManagerProxy = PuzzleManagerProxy(self.puzzleManager)
        self.inputManager = InputManager(self.puzzleManagerProxy)

    def startGame(self):
        self.puzzleManagerProxy.spawnBlock()
        self.puzzleManagerProxy.spawnBlock()

        while self.isGameOver == False:
            self.printText.printGrid()
            self.inputManager.getKey()
            self.isGameOver = self.puzzleManagerProxy.isGameOver()

        self.printText.printGrid()
        self.printText.gameOver()

# =========================
# Main
# =========================

def main():
    gameManager = GameManager()
    gameManager.startGame()

main()
````

````python
from ast import match_case

# =========================
# Calculator
# =========================

class Calculator():
  def calculate(self):
    pass

class Plus(Calculator):
  def calculate(self,a,b):
    return a+b

class Minus(Calculator):
  def calculate(self,a,b):
    return a-b

class Multiply(Calculator):
  def calculate(self,a,b):
    return a*b

class Divide(Calculator):
  def calculate(self,a,b):
    return a/b

# =========================
# calculateStrategy
# =========================

def calculateStrategy(calculateType)->Calculator:
  match calculateType:
    case '+':
      return Plus()
    case '-':
      return Minus()
    case '*':
      return Multiply()
    case '/':
      return Divide()
    case _:
      print("error")
      return

# =========================
# main
# =========================

def main():
  clType=input("input calculate type (+ - * /): ")
  calculator=calculateStrategy(clType)
  result=calculator.calculate(20,10)
  print(result)

main()
````

````python
from abc import ABC, abstractmethod

# =========================
# Command
# =========================

#Command: ëª…ë ¹ ìº¡ìŠí™”
class Command(ABC):
  @abstractmethod
  def execute(self):
    pass

#Recevier: ì‹¤ì œë¡œ ì¼ì„ í•˜ëŠ” ê°ì²´
class Magic():
  def healMagic(self):
    print("Heal 10 HP!")

  def atkBuffMagic(self):
    print("Attack buff to team!")

  def fireMagic(self):
    print("Deal 5 fire damage!")

  def iceMagic(self):
    print("Deal 5 ice damage!")

#Concrete Command: ì‹¤ì²´í™”
class DealMagic(Command):
  def __init__(self,magic:Magic,commandList):
    self.magic=magic
    self.commandList=commandList

  def execute(self):
    for command in self.commandList:
      if command=='fireMagic':
        self.magic.fireMagic()
      elif command=='iceMagic':
        self.magic.iceMagic()

class HealMagic(Command):
  def __init__(self,magic:Magic,commandList):
    self.magic=magic
    self.commandList=commandList

  def execute(self):
    for command in self.commandList:
      if command=='healMagic':
        self.magic.healMagic()
      elif command=='atkBuffMagic':
        self.magic.atkBuffMagic()

# =========================
# MagicController
# =========================

#Invoker: ëª…ë ¹ ì‹¤í–‰ ê°ì²´
class MagicController():
  def __init__(self):
    self.commandList=[]

  def addCommand(self,command):
    self.commandList.append(command)

  def executeCommands(self):
    for command in self.commandList:
      command.execute()

# =========================
# main
# =========================

def main():
  magic=Magic()
  magicController=MagicController()

  healMagic=HealMagic(magic,['healMagic','atkBuffMagic','healMagic'])
  magicController.addCommand(healMagic)

  dealMagic=DealMagic(magic,['fireMagic','iceMagic','fireMagic'])
  magicController.addCommand(dealMagic)

  magicController.executeCommands()

main()
````

````python
# =========================
# FilterHandler
# =========================

#Handler
class FilterHandler():
  def __init__(self):
    self.nextHandler=None

  def setNextHandler(self,handler):
    self.nextHandler=handler

  def handle(self,message):
    if self.nextHandler:
      return self.nextHandler.handle(message)

    print("Message is not spam!")
    return None

# =========================
# ConcreteHandler
# =========================

class CheckAd(FilterHandler):
  def handle(self,message):
    if message=='ad':
      print("Blocked AD message!")
    else:
      super().handle(message)

class CheckSpam(FilterHandler):
  def handle(self,message):
    if message=='unknown':
      print("Defer unknown message!") #ë³´ë¥˜í•˜ë‹¤
    else:
      super().handle(message)

class CheckBlocked(FilterHandler):
  def handle(self,message):
    if message=='blocked':
      print("Defer blocked message!")
    else:
      super().handle(message)

# =========================
# main
# =========================

def main():
  checkAd=CheckAd()
  checkSpam=CheckSpam()
  checkBlocked=CheckBlocked()

  checkAd.setNextHandler(checkSpam)
  checkSpam.setNextHandler(checkBlocked)

  checkAd.handle('blocked')
  checkAd.handle('ad')
  checkAd.handle('friend')

main()
````

````python
from random import *
import time

# =========================
# Observer
# =========================

class Observer():
    def update(self, subject):
      pass

class TakeDamage(Observer):
    def update(self, subject):
      print(f"[{subject.name}]ì˜ í˜„ìž¬ HP: {subject.hp}")

      if subject.hp <= 0:
        print(f"{subject.name}ê°€ ì“°ëŸ¬ì¡ŒìŠµë‹ˆë‹¤!")
        subject.isAlive = False

# =========================
# Subject
# =========================

class Subject():
    def __init__(self):
      self.observerList = []

    def register(self, observer: Observer):
      self.observerList.append(observer)

    def notify(self):
      for observer in self.observerList:
        observer.update(self)

# =========================
# Character
# =========================

class Character(Subject):
    def __init__(self):
      super().__init__()
      self.hp = 0
      self.name=None
      self.isAlive=True

    def takeDamage(self, damage):
      self.hp -= damage
      if self.hp < 0:
        self.hp = 0
      self.notify()


class Player(Character):
    def __init__(self):
      super().__init__()
      self.hp=randint(10,15)
      self.name="Knight"


class Enemy(Character):
    def __init__(self):
      super().__init__()
      self.hp=randint(10,15)
      self.name="Goblin"

# =========================
# GameManager
# =========================

class GameManager():
    def __init__(self):
        self.player = Player()
        self.enemy = Enemy()

        self.playerObserver = TakeDamage()
        self.enemyObserver = TakeDamage()
        self.player.register(self.playerObserver)
        self.enemy.register(self.enemyObserver)

    def run(self):
      print(f"{self.player.name} hp: {self.player.hp}")
      print(f"{self.enemy.name} hp: {self.enemy.hp}")
      print("\n"+"-"*30)

      while self.player.isAlive and self.enemy.isAlive:

        damage = randint(1, 5)
        print(f"player attack {damage} damage!")
        self.enemy.takeDamage(damage)
        print("\n")

        time.sleep(1)

        damage = randint(1, 5)
        print(f"enemy attack {damage} damage!")
        self.player.takeDamage(damage)
        print("\n"+"="*30)

        time.sleep(1)

      if self.player.hp <= 0:
        print("Game Over!")
      else:
        print("Win!")

# =========================
# main
# =========================
def main():
    gameManager = GameManager()
    gameManager.run()

main()
````

````python
# =========================
# InventoryIterator
# =========================

class InventoryIterator():
  def __init__(self,inventory):
    self.itemList=inventory.itemList
    self.pos=-1

  def hasNext(self):
    if self.pos<len(self.itemList)-1:
      return True
    return False

  def next(self):
    if self.hasNext():
      self.pos+=1
      return self.itemList[self.pos]

# =========================
# Inventory
# =========================

#aggregate: ë°ì´í„°ë¥¼ ë³´ê´€í•˜ê³  ë°˜ë³µìžë¥¼ ë°˜í™˜í•˜ëŠ” í´ëž˜ìŠ¤
class Inventory():
  def __init__(self):
    self.itemList=[]

  def add(self,item,itemClass):
    self.itemList.append((item,itemClass))

  def getIterator(self):
    return InventoryIterator(self)

# =========================
# printItemInfo
# =========================

def printItemInfo(iterator):
  while iterator.hasNext():
    print(iterator.next())
  print(' ')

# =========================
# main
# =========================

def main():
  inventory=Inventory()

  inventory.add('short sword','common')
  inventory.add('long sword','common')
  inventory.add('fire sword','rare')

  iterator=inventory.getIterator()
  printItemInfo(iterator)

main()
````
