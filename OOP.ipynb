````python
from os import X_OK
from ast import match_case
import time
from random import *
from google.colab import output
from abc import ABC, abstractmethod



# =========================
# Character
# =========================

class Character():
  def __init__(self,health):
    self.health=health

  def isAlive(self):
    return self.health>0

  def takeDamage(self,damage):
    self.health-=damage
    if self.health<0:
      self.health=0


class Player(Character):
  def __init__(self):
    super().__init__(health=15)

class Enemy(Character):
  def __init__(self):
    super().__init__(health=randint(10,20))



# =========================
# AttackAction
# =========================


class AttackActionAbstract(ABC):
  @abstractmethod
  def playAction(self,enemy):
    pass


class AttackAction(AttackActionAbstract):
  def playAction(self,enemy):
    damage=randint(1,3)
    enemy.takeDamage(damage)
    print(f"Player attack {damage} damage!")


# =========================
# DefenseAction
# =========================



class DefenseActionAbstract(ABC):
  @abstractmethod
  def defenseAction(self,player,enemy,damage):
    pass


class CounterDefense(DefenseActionAbstract):
  def defenseAction(self,player,enemy,enemyDamage):
    blockAmount=randint(0,2)
    print(f"Player take {enemyDamage} damage and block {blockAmount} damage!")
    enemyDamage=max(0,enemyDamage-blockAmount)
    player.takeDamage(enemyDamage)
    time.sleep(1)

    if(player.isAlive()):
      playerDamage=randint(1,2)
      enemy.takeDamage(playerDamage)
      print(f"Player counter {playerDamage} damage!")

class AvoidDefense(DefenseActionAbstract):
  def defenseAction(self,player,enemy,damage):
    isAvoid=randint(0,1)
    if(isAvoid==1):
      damage=2
      print(f"Avoid success! and attack {damage} damage!")
      enemy.takeDamage(damage)
    else:
      print(f"Avoid fail! and take {damage} damage!")
      player.takeDamage(damage)


# =========================
# InputManager
# =========================


class InputManager():

  def __init__(self):
    self.key=""

  def getKey(self):
    self.key=input().upper()
    return self.key

  def checkKey(self,key):
    match self.key:
      case 'Z':
        return "attack"
      case 'X':
        return "counter"
      case 'C':
        return "avoid"
      case _:
        print("Error: please enter the correct key")
        return "-1"

# =========================
# PrintText
# =========================

class PrintText():
  def printInfo(self,enemy,player):
    print("\n\n\n")
    print(f"Enemy Health: {enemy.health}, Player Health: {player.health}")
    print("attack: Z(1 ~ 3)   counter: X(0 ~ 2 block, 1~2 attack)   avoid: C(50% avoid and 2 attack)")
    print("-"*30)

  def printResult(self,isWin):
    if(isWin==True):
      print("Win!")
    else:
      print("Game Over!")

# =========================
# checkAction()
# =========================

class CheckAction():
  def __init__(self,player,enemy):
    self.player=player
    self.enemy=enemy

    self.attackType={
    "attack":AttackAction(),
    }

    self.defenseType={
    "counter":CounterDefense(),
    "avoid":AvoidDefense()
    }

  def isAttack(self,actionType):
    if(actionType in self.attackType):
      attackAction=self.attackType[actionType]

      attackAction.playAction(self.enemy)
      time.sleep(1)

  def isDefense(self,actionType):
    damage=randint(1,3)
    if(actionType in self.defenseType):
      defenseAction=self.defenseType[actionType]
      defenseAction.defenseAction(self.player,self.enemy,damage)
    else:
      self.player.takeDamage(damage)
      print(f"Player take {damage} damage!")


# =========================
# GameManager
# =========================


class GameManager():
  def __init__(self):
    self.player=Player()
    self.enemy=Enemy()
    self.printText=PrintText()
    self.inputManager=InputManager()
    self.checkAction=CheckAction(self.player,self.enemy)

  def playerTurn(self):
    key="-1"
    while key== "-1":
      self.printText.printInfo(self.enemy,self.player)
      key=self.inputManager.checkKey(self.inputManager.getKey())

    self.checkAction.isAttack(key)
    if(self.enemy.isAlive()):
      self.checkAction.isDefense(key)


  def run(self):
    while self.player.isAlive() and self.enemy.isAlive():
      self.playerTurn()

    if self.player.isAlive():
      isWin=True
      self.printText.printResult(isWin)
    else:
      isWin=False
      self.printText.printResult(isWin)

# =========================
# start
# ====================z=====

def start():
  game=GameManager()
  game.run()

start()
````

````python
import copy
from random import *
from abc import ABC, abstractmethod
myId=0

# =========================
# Singleton
# =========================

class SingletonId():
  _instance=None

  def __new__(cls,id):
    if not cls._instance:
      cls._instance=super(SingletonId,cls).__new__(cls)
      cls._instance.id=id
    print(f"your id is {cls._instance.id}")
    return cls._instance

# =========================
# Factory Method
# =========================

class CharacterClass():
  def description(self):
    pass

class Knight(CharacterClass):
  def description(self):
    print("Have strong sword")

class Archer(CharacterClass):
  def description(self):
    print("Have long bow")


class CharacterClassFactory():
  def chooseClass(self):
    pass

class KnightFactory(CharacterClassFactory):
  def chooseClass(self):
    self.getSword()
    SingletonId(myId)
    return Knight()

  def getSword(self):
    print("get greate sword")

class ArcherFactory(CharacterClassFactory):
  def chooseClass(self):
    self.getCape()
    return Archer()

  def getCape(self):
    print("get cool cape")

# =========================
# Abstract Factory
# =========================

class Templete(ABC):
  @abstractmethod
  def setCharacter():
    pass

class ChildrenTemplete(Templete):
  def setCharacter():
    character=CharacterBuilder().setName("Rose").setGender("female").setHeight(80).setWeight(30).setHairStyle("long hair").setAge(10).build()
    return character

class ElderTemplete(Templete):
  def setCharacter():
    character=CharacterBuilder().setName("Kain").setGender("male").setHeight(160).setWeight(60).setHairStyle("None").setAge(80).build()
    return character


class TempleteFactory(ABC):
  @abstractmethod
  def getTemplete(self):
    pass

class ChildrenFactory(TempleteFactory):
  def getTemplete(self):
    character=ChildrenTemplete.setCharacter()
    return character

class ElderFactory(TempleteFactory):
  def getTemplete(self):
    character=ElderTemplete.setCharacter()
    return character

# =========================
# Builder
# =========================

class Character():
  def __init__(self,name,gender,height,weight,hairStyle,age):
    self.name=name
    self.gender=gender
    self.height=height
    self.weight=weight
    self.hairStyle=hairStyle
    self.age=age

  #prototype
  def clone(self):
    return copy.deepcopy(self)

  def printInfo(self):
    print(f"name: {self.name}, gender: {self.gender}, height: {self.height}, weight: {self.weight}, hair: {self.hairStyle}. age: {self.age}")

class CharacterBuilder():
  def __init__(self):
    self.name=None
    self.gender = None
    self.height = None
    self.weight = None
    self.hairStyle = None
    self.age=None

  def setName(self,name):
    self.name=name
    return self


  def setGender(self,gender):
    self.gender=gender
    return self

  def setHeight(self,height):
    self.height=height
    return self

  def setWeight(self,weight):
    self.weight=weight
    return self

  def setHairStyle(self,hairStyle):
    self.hairStyle=hairStyle
    return self

  def setAge(self,age):
    self.age=age
    return self

  def build(self):
    character=Character(self.name,self.gender,self.height,self.weight,self.hairStyle,self.age)
    return character


class MaleCharacterBuilder(CharacterBuilder):
  def __init__(self):
    super().__init__()
    self.gender="male"

class FemaleCharacterBuilder(CharacterBuilder):
  def __init__(self):
    super().__init__()
    self.gender="female"

# =========================
# main
# =========================

def main():

  myId=randint(1,1000000)
  anotherId=randint(1,1000000)
  SingletonId(myId)
  SingletonId(anotherId)
  print("\n")


  characterBuilder=CharacterBuilder()
  character=characterBuilder.setName("July").setGender("female").setHeight(160).setWeight(50).setHairStyle("long hair").setAge(23).build()
  character.printInfo()

  maleCharacter=MaleCharacterBuilder().setName("Jake").setHeight(180).setWeight(80).setHairStyle("short hair").setAge(20).build()
  maleCharacter.printInfo()
  print("\n")


  characterClone=character.clone()
  characterClone.name="clone"
  characterClone.printInfo()
  print("\n")


  childrenFactory=ChildrenFactory()
  childrenCharacter=childrenFactory.getTemplete()
  childrenCharacter.printInfo()

  elderFactory=ElderFactory()
  elderCharacter=elderFactory.getTemplete()
  elderCharacter.printInfo()
  print("\n")

  knight=KnightFactory().chooseClass()
  knight.description()
  archer=ArcherFactory().chooseClass()
  archer.description()
  print("\n")

main()
````
